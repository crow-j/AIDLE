#### 8. 대량메일 발송 여부 확인
관련 헤더
- /header/header/mail_from
- /header/header/date

관련 컬럼 
- mail_size

```python
eml_df = pd.read_csv("../../parsed_eml_3.csv")
```

```python
eml_df['file_size']
```

```python
eml_df['mail_from_and_date'] = eml_df["/header/header/mail_from"].str.cat(eml_df["/header/header/date"])
```

```python
eml_df['mail_from_and_date']
```

```python
# 동일한 송신자가 동시간에 보낸 메일의 개수 확인
plt.figure(figsize=(20, 8))
plt.title('Origin Sender IP address', fontsize=20)
sns.countplot(x='mail_from_and_date', data=eml_df)
plt.xticks(rotation=30, ha='right')
plt.show()
```

```python
# 동시간대에 보내진 메일 개수 확인
plt.figure(figsize=(20, 8))
plt.title('Origin Sender IP address', fontsize=20)
sns.countplot(x='/header/header/date', data=eml_df)
plt.xticks(rotation=30, ha='right')
plt.show()
```

```python
plt.figure(figsize=(30, 40))
#plt.title('', fontsize=20)
sns.scatterplot(data=eml_df, x="/header/header/mail_from", y="/header/header/date", hue="file_size")
plt.xticks(rotation=30, ha='right')
plt.show()
```

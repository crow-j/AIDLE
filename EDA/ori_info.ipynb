## 최초 발신자 IP, server와 최초 수신자 메일주소 부분 EDA

eml 파일 내 제일 아래 Received 필더 값 중 from/by/for에 해당하는 값이 각각 최초 발신자 IP, 최초 발신자 server, 최초 수신자 메일주소라고 판단

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
eml_df = pd.read_csv("../../parsed_eml_3.csv", parse_dates=["/header/date", "/header/header/date", "/header/received/0/src_at", 
                                               "/header/received/1/date", "/header/received/2/date", 
                                               "/header/received/3/date", "/header/received/4/date", "/header/received/5/date"],
                      date_parser=lambda col: pd.to_datetime(col, utc=True))
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x='ori_sender_ip', data=eml_df)
plt.xticks(rotation=30, ha='right')
plt.title('Origin sender IP')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x='ori_sender_server', data=eml_df)
plt.xticks(rotation=30, ha='right')
plt.title('Origin sender Server')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x='ori_receiver_mail', data=eml_df)
plt.xticks(rotation=30, ha='right')
plt.title('Origin receiver mail address')
plt.show()
```

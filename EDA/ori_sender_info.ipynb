## 발신자 IP, server, mail주소 부분 EDA

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
eml_df = pd.read_csv("add_origin_received.csv", parse_dates=["/header/date", "/header/received/0/src_at", 
                                               "/header/received/1/date", "/header/received/2/date", 
                                               "/header/received/3/date", "/header/received/4/date", "/header/received/5/date"],
                      date_parser=lambda col: pd.to_datetime(col, utc=True))
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x='ori_sender_ip', data=eml_df)
plt.xticks(rotation=30, ha='right')
plt.title('Origin sender IP')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x='ori_sender_server', data=eml_df)
plt.xticks(rotation=30, ha='right')
plt.title('Origin sender Server')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x='ori_sender_mail', data=eml_df)
plt.xticks(rotation=30, ha='right')
plt.title('Origin sender mail address')
plt.show()
```

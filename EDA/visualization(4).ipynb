#### 4. 송신 호스트 IP별로 시각화
관련 헤더
- /header/received/n/from: 최초/중간/최종 송신 호스트 IP
    * n == last_num일 때, 최초 송신자 IP정보
        * ori_sender_ip에 저장되어 있음(find_origin_info.ipynb 에서 처리)
            * IP가 아닌 값(메일 서버 주소로 보임)이 존재 **=> 어떻게 처리하지**
            * 여러개의 IP가 존재하는 경우도 있음 **=> 어떻게 처리하지**
    * n == 0일 때, 최종 송신 호스트 IP 정보

```python
eml_df = pd.read_csv("../../parsed_eml_3.csv")
```

```python
# 최초 송신자 IP 정보
plt.figure(figsize=(20, 8))
plt.title('Origin Sender IP address', fontsize=20)
sns.countplot(x="ori_sender_ip", data=eml_df)
plt.xticks(rotation=30, ha='right')
plt.show()
```

```python
# 최종 송신 호스트 IP 주소
plt.figure(figsize=(30, 6))
plt.title('Last Sender IP address', fontsize=20)
sns.countplot(x="/header/received/0/from", data=eml_df)
plt.xticks(rotation=30, ha='right')
plt.show()
```

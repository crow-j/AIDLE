## 특징별로 시각화(5)

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

#### 5. 최초 송신자 도메인별로 시각화
관련 헤더
- /header/header/mail_from: 최초 송신자 메일 주소

```python
eml_df = pd.read_csv("../../parsed_eml_3.csv")
```

```python
eml_df['mail_from_domain'] = eml_df['/header/header/mail_from'].str.split('@', expand=True)[1]
```

```python
eml_df['mail_from_domain']
```

```python
# 최초 송신자 도메인 정보
plt.figure(figsize=(20, 8))
plt.title('Origin Sender Domain', fontsize=20)
sns.countplot(x="mail_from_domain", data=eml_df)
plt.xticks(rotation=30, ha='right')
plt.show()
```

#### 6. 수신 도메인별로 시각화
관련 헤더
- /header/to: 수신자의 메일 주소

```python
eml_df['to_domain'] = eml_df['/header/to'].str.split('@', expand=True)[1]
```

```python
# 수신자 도메인 정보
plt.figure(figsize=(20, 8))
plt.title('Last Receiver Domain', fontsize=20)
sns.countplot(x="to_domain", data=eml_df)
plt.xticks(rotation=30, ha='right')
plt.show()
```
